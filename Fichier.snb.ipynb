```scala
val rdd2 = rdd
  .flatMap(s => s.split(' '))
  .filter(s => s != " " && s != "")
  .map(s => (s, 1))
  .reduceByKey(_ + _)
  .groupBy  .sortBy(e => e._2, false)
```

```scala
rdd2.map(v => s"${v._1},${v._2}").saveAsTextFile("/home/ekoffi/Downloads/result.csv")
```
