```scala
val df = sparkSession.read.text("/home/ekoffi/Downloads/SampleTextFile_1000kb.txt")
```

```scala
df.printSchema()
```

```scala
val df2 = df
  .flatMap(a => a.mkString.split(' '))
  .filter(v => v.nonEmpty)
```

```scala
df2.printSchema()
```

```scala
val df3 = df2
  .groupBy($"value")
  .count()
  .sort(desc("count"))
```

```scala
df3.printSchema()
```

```scala
df3
  .write
  .mode("overwrite")
  .csv("/home/ekoffi/Downloads/result.csv")
```

```scala
val replaceSed = udf((s: String) => s.replace("sed", "awk"))
```

```scala
val total = df2.count()
```

```scala
import org.apache.spark.sql.types.StringType

df3
.withColumn("value2", $"value")
.select(replaceSed($"value2"), $"count", $"count" * 100.0 / total as "densite" cast StringType)
```
